```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset

```python
df = pd.read_csv('/content/advertising.csv')
df.head(15)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
df['TV'].corr(df['Sales'])
```

```python
df['TV'].corr(df['Sales'])
plt.scatter(df['TV'], df['Sales'])
plt.show()
```

```python
df['TV'].corr(df['Sales'])
plt.hist(df['TV']) # Removed the second argument to use default binning
plt.xlabel('Radio')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Radio'].corr(df['Sales'])
plt.hist(df['Radio'], bins=10, color="green")
plt.xlabel('Radio')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Newspaper'].corr(df['Sales'])
plt.hist(df['Newspaper'], bins=10, color="red")
plt.xlabel('Newspaper')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

using linear regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['TV']], df['Sales'], test_size=0.2, random_state=2)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
res=model.predict(x_test)
print(res)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.coef_*69.2+model.intercept_
```

```python
plt.plot(res)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, model.predict(x_train),'r')
plt.show()
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_train, model.predict(x_train),'r')
plt.show()
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test,7.03259359+0.04753664*x_test,'r')
plt.show()
```
